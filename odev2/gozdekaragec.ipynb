```python
# Akıllı Gün Planlayıcı
# Ödev 2

while True:
    uyku = float(input("Günlük uyku sürenizi giriniz (saat): "))
    ders = float(input("Günlük ders sürenizi giriniz (saat): "))
    serbest = float(input("Günlük serbest zamanınızı giriniz (saat): "))
    if uyku < 0 or ders < 0 or serbest < 0:
        print("Hatalı giriş! Süreler negatif olamaz.\n")
        continue
    if uyku + ders + serbest > 24:
        print("Hatalı giriş! Sürelerin toplamı 24 saati geçemez.\n")
        continue
    break
if uyku < 6:
    uyku_durumu = "Yetersiz uyku"
elif uyku <= 8:
    uyku_durumu = "Normal uyku"
else:
    uyku_durumu = "Fazla uyku"
if ders >= 6:
    calisma_durumu = "Yoğun çalışma günü"
else:
    calisma_durumu = "Normal çalışma günü"
if uyku < 6 and ders >= 6:
    genel_durum = "Verimsiz gün"
else:
    genel_durum = "Dengeli gün"
print("\n--- Günlük Değerlendirme ---")
print("Uyku durumu:", uyku_durumu)
print("Çalışma durumu:", calisma_durumu)
print("Genel değerlendirme:", genel_durum)
```
